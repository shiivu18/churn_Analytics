```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Churn Value', data=df)

plt.title("Customer Churn Distribution")
plt.show()
```

```python
df.columns.tolist()






sns.countplot(x='Contract', hue='Churn Value', data=df)

plt.title("Contract Type vs Churn")
plt.xticks(rotation=15)
plt.show()
```

```python
sns.histplot(data=df, x='Tenure Months', hue='Churn Value', bins=30)

plt.title("Tenure Distribution by Churn")
plt.show()
```

### Insight

Customers with month-to-month contracts appear to churn significantly more than customers with long-term contracts.

```python
sns.countplot(x='Internet Service', hue='Churn Value', data=df)

plt.title("Internet Service vs Churn")
plt.xticks(rotation=15)
plt.show()
```

Customers using certain internet services may show higher churn rates due to pricing, service quality, or competition.

```python
sns.countplot(x='Senior Citizen', hue='Churn Value', data=df)

plt.title("Senior Citizens vs Churn")
plt.show()
```

Age-related customer groups may have different retention behavior.

```python
plt.figure(figsize=(10,5))

sns.countplot(x='Payment Method', hue='Churn Value', data=df)

plt.title("Payment Method vs Churn")
plt.xticks(rotation=45)
plt.show()
```

Electronic check users often show higher churn behavior in many telecom datasets.

```python
sns.countplot(x='Gender', hue='Churn Value', data=df)

plt.title("Gender vs Churn")
plt.show()
```

Gender may or may not significantly influence churn.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Tech Support', hue='Churn Value', data=df)

plt.title("Tech Support vs Churn")
plt.xticks(rotation=15)
plt.show()
```

Customers without tech support services may churn more frequently.
